```python
k = 50_000
base_loan = 1_000
lending_rate = 0.055
```

```python
def compute_lending(X) : 
    x1, x2, x3 = X[0], X[1], X[2]
    xi = np.random.uniform(0, 1, size = k)
    default = y2(xi, x1, x2, x3)
    mean_default = np.mean(default)
    print(mean_default)
    expected_profit = (1 - mean_default) * base_loan * (1 + lending_rate) - base_loan
    return expected_profit

loans = np.apply_along_axis(compute_lending, axis=1, arr=test_set)
```
